# wine-quality analysis using RandomForest

```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the Data
data=pd.read_csv('winequalityN.csv')
data.shape
```

```python
data.head()
```

```python
# Encoding the type
data.type=data.type.apply(lambda x: 0 if x=='white' else 1)
data.type
```

```python
# checking null values
data.isnull().sum()
```

```python
data['fixed acidity']=data['fixed acidity'].fillna(data['fixed acidity'].mean())
data['volatile acidity']=data['volatile acidity'].fillna(data['volatile acidity'].mean())
data['citric acid']=data['citric acid'].fillna(data['citric acid'].mean())
data['residual sugar']=data['residual sugar'].fillna(data['residual sugar'].mean())
data['chlorides']=data['chlorides'].fillna(data['chlorides'].mean())
data['pH']=data['pH'].fillna(data['pH'].mean())
data['sulphates']=data['sulphates'].fillna(data['sulphates'].mean())
data.isnull().sum()
```

```python
data.columns
```

```python
plt.figure(figsize=(20,20))
pltnum=1

for columns in data:
    if pltnum <=12:
        ax=plt.subplot(3,4,pltnum)
        sns.distplot(data[columns])
    pltnum=pltnum+1
```

```python
plt.figure(figsize=(20,20))
pltnum=1
for columns in data:
    if pltnum <= 12:
        ax=plt.subplot(4,3,pltnum)
        sns.boxplot(x=data[columns])
    pltnum=pltnum+1
```

```python
# there are some outliers in the dataset lets remove them
for columns in ['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol',
       'quality']:
       data=data[data[columns]< data[columns].quantile(0.90)]
    
```

```python
plt.figure(figsize=(20,20))
pltnum=1
for columns in data:
    if pltnum <= 12:
        ax=plt.subplot(4,3,pltnum)
        sns.boxplot(x=data[columns])
    pltnum=pltnum+1
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

# Doing Eda with Dependent and Independent varables


```python
for columns in ['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']:


        fig,ax=plt.subplots(1,3,figsize=(20,5))
        sns.barplot(data=data,x='quality',y=data[columns],ax=ax[0])
        sns.lineplot(data=data,x='quality',y=data[columns],ax=ax[1])
        sns.violinplot(data=data,x='quality',y=data[columns],ax=ax[2])
    
```

```python
x=data.drop(['quality'],axis=1)
y=data['quality']
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test =train_test_split(x,y,test_size=.1,random_state=3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier()
```

```python
# fitting the data into the model
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
# now chcking the the accuracy of the model
acc=accuracy_score(pred,y_test)
```

```python
print("The accuracy of the model ",acc)
```
